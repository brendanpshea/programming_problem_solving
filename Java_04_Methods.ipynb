## Why Methods?

Imagine Bob Belcher had to explain every single step of making a burger to his family every single time:

*"Okay, first get the beef patty from the fridge. Now put it on the grill. Wait 3 minutes. Flip it. Wait 3 more minutes. Get a bun. Toast the bun. Put lettuce on the bun. Put tomato on the bun. Put the patty on the bun. Add the Burger of the Day toppings..."*

This would be exhausting! Instead, Bob has **recipes**‚Äînamed procedures that everyone knows how to follow. He can just say "Make me a Burger of the Day" and the whole process happens.

**Methods are like recipes for your code.** They let you:

- **Reuse code** ‚Äî Write once, use many times
- **Organize** ‚Äî Break complex problems into manageable pieces
- **Abstract** ‚Äî Hide details behind a simple name
- **Debug** ‚Äî Fix one method, fix it everywhere it's used

> **The DRY Principle:** Don't Repeat Yourself. If you're copying and pasting code, you probably need a method.

## What is a Method?

A **method** is a named, reusable block of code that performs a specific task.

Think of it like a recipe card in Bob's kitchen:

| Recipe Card | Method |
|-------------|--------|
| Recipe name | Method name |
| Ingredients needed | Parameters (inputs) |
| Instructions | Method body (code) |
| Finished dish | Return value (output) |

Some recipes produce something (a burger), while others just perform an action (cleaning the grill). Similarly, some methods return a value, while others just do something.

You've already been using methods! `System.out.println()` is a method. `Math.sqrt()` is a method. Now you'll learn to write your own.

---

*Let's start by learning the basic structure of a method.*

## Method Basics

Here's the structure of a Java method:

```
public static returnType methodName(parameters) {
    // method body - the code that runs
}
```

Let's break down each part with a Bob's Burgers analogy:

| Part | Example | Analogy |
|------|---------|--------|
| `public static` | `public static` | Recipe is available to everyone in the kitchen (for now, always use this) |
| `returnType` | `void`, `int`, `String` | What the recipe produces: nothing (`void`), a number, text, etc. |
| `methodName` | `makeBurger` | The name you call to run this recipe |
| `parameters` | `(String burgerType)` | Ingredients/information needed |
| `{ body }` | `{ ... }` | The actual instructions |

Method names follow the same rules as variables: start with lowercase, use camelCase, be descriptive. Good names read like commands: `calculateTotal`, `printReceipt`, `isOpen`.

### Calling Methods

Defining a method is like writing a recipe card. **Calling** (or invoking) a method is like actually following that recipe.

To call a method, use its name followed by parentheses:

```java
greetCustomer();  // Call the method
greetCustomer();  // Call it again!
greetCustomer();  // And again!
```

Each call executes all the code inside the method. Three calls = three greetings printed.

**Common mistake:** Forgetting the parentheses. `greetCustomer` without `()` doesn't call the method‚Äîit's just a reference to it (which causes an error in our context).

**Key insight:** The method definition happens once, but you can call it as many times as you need. Write the code once, use it everywhere.

**Full Program #1: BobsGreeting.java** demonstrates method reuse. Notice:
- The `greetCustomer()` method is defined once, outside of `main`
- We call it three times from `main`, and it runs each time
- If we wanted to change the greeting, we'd only change it in one place!

````python
%%writefile BobsGreeting.java
public class BobsGreeting {

    // Method definition - the "recipe"
    public static void greetCustomer() {
        System.out.println("Welcome to Bob's Burgers!");
        System.out.println("Home of the Burger of the Day!");
        System.out.println("Free French Fries Today!");
        System.out.println();
    }

    public static void main(String[] args) {
        System.out.println("=== BREAKFAST RUSH ===");
        greetCustomer();  // First call

        System.out.println("=== LUNCH RUSH ===");
        greetCustomer();  // Second call

        System.out.println("=== DINNER RUSH ===");
        greetCustomer();  // Third call
    }
}
````

### Where to Define Methods

Methods must be defined **inside the class** but **outside of `main`** (and outside of other methods).

```java
public class Restaurant {
    
    // ‚úì CORRECT: Method defined inside class, outside main
    public static void greetCustomer() {
        System.out.println("Welcome!");
    }
    
    public static void main(String[] args) {
        greetCustomer();  // Call the method
        
        // ‚úó WRONG: Can't define a method inside main!
        // public static void badMethod() { }  // ERROR!
    }
}
```

The order of methods in the file doesn't matter‚ÄîJava finds them before running. You can put `main` first or last, and helper methods anywhere.

### The `static` Keyword (For Now)

You might wonder why we write `public static` on every method. Here's a brief explanation:

- **`static`** means the method belongs to the *class itself*, not to an individual object
- Since `main` is `static`, it can only directly call other `static` methods
- For now, **always make your methods `public static`**

When we learn about Object-Oriented Programming, we'll explore non-static methods that belong to objects. Think of it like this:

- **Static methods** = Restaurant-wide procedures ("How to close up the restaurant")
- **Non-static methods** = Individual actions ("Bob flips a burger" vs "Tina flips a burger")

For now, just remember: `public static` goes before the return type.

## Parameters ‚Äî Giving Methods Information


Our `greetCustomer()` method always does the exact same thing. But what if we want to personalize the greeting? What if Bob wants to greet customers by name?

Consider the difference:
- "Make me a burger" ‚Äî Which burger? How many? This is too vague.
- "Make me a Burger of the Day" ‚Äî Now Bob knows exactly what to do!

**Parameters** are variables that let you pass information into a method. They're like the specific ingredients or instructions a recipe needs.

Without parameters, every method call is identical. With parameters, methods become flexible and useful in many situations.

### Defining Parameters

Parameters go inside the parentheses, with a type and a name:

```java
public static void greetCustomer(String name) {
    System.out.println("Welcome to Bob's Burgers, " + name + "!");
    System.out.println("What can I get for you today?");
}
```

Now when we call the method, we must provide a value:

```java
greetCustomer("Teddy");      // Welcome to Bob's Burgers, Teddy!
greetCustomer("Mort");       // Welcome to Bob's Burgers, Mort!
greetCustomer("Mr. Fischoeder");  // Welcome to Bob's Burgers, Mr. Fischoeder!
```

The parameter `name` acts like a variable inside the method. Each time we call the method with a different value, `name` takes on that value.

### Arguments vs Parameters

These two terms are often confused, but there's a distinction:

| Term | Where | What It Is | Example |
|------|-------|------------|--------|
| **Parameter** | Method definition | The variable that receives data | `String name` in `void greet(String name)` |
| **Argument** | Method call | The actual value passed in | `"Teddy"` in `greet("Teddy")` |

Think of it like ordering at Bob's:
- The **parameter** is the blank on the order pad: "Burger type: _____"
- The **argument** is what you write in: "Burger of the Day"

In practice, many people use these words interchangeably, but knowing the difference helps when reading documentation or error messages.

### Multiple Parameters

Methods can have multiple parameters, separated by commas:

```java
public static void prepareBurger(String burgerName, boolean addFries) {
    System.out.println("Preparing: " + burgerName);
    if (addFries) {
        System.out.println("Adding a side of fries!");
    }
}
```

When calling, provide arguments in the **same order** as the parameters:

```java
prepareBurger("Rest In Peas Burger", true);   // Correct!
prepareBurger(true, "Rest In Peas Burger");   // ERROR! Wrong order!
```

**Order matters!** Java matches arguments to parameters by position, not by name. The first argument goes to the first parameter, second to second, and so on.

**Full Program #2: BurgerOrder.java** shows a method with three parameters of different types:
- `String customer` ‚Äî who's ordering
- `String burger` ‚Äî what they want
- `int quantity` ‚Äî how many

Each call provides different arguments, but the method handles them all the same way. That's the power of parameters!

````python
%%writefile BurgerOrder.java
public class BurgerOrder {

    public static void takeOrder(String customer, String burger, int quantity) {
        System.out.println("Order for: " + customer);
        System.out.println("  Burger: " + burger);
        System.out.println("  Quantity: " + quantity);
        System.out.println("  " + customer + ", your order is being prepared!");
        System.out.println();
    }

    public static void main(String[] args) {
        // The Belcher family places their orders
        takeOrder("Bob", "New Bacon-ings Burger", 1);
        takeOrder("Linda", "Foot Feta-ish Burger", 1);
        takeOrder("Tina", "A Good Manchego is Hard to Find Burger", 2);
        takeOrder("Gene", "Poutine on the Ritz Burger", 3);
        takeOrder("Louise", "Shoot-Out at the OK-ra Corral Burger", 1);
    }
}
````

### The `return` Statement

To send a value back from a method:
1. Declare the **return type** (instead of `void`)
2. Use the **`return` statement** to send back a value

```java
public static int countPatties(int numberOfBurgers) {
    int patties = numberOfBurgers * 2;  // Double patties on each burger!
    return patties;  // Send this value back to the caller
}
```

Key rules:
- The return type (`int`) must match what you actually return
- `return` immediately exits the method
- Every path through the method must return a value (for non-void methods)

You can also return expressions directly:

```java
public static int countPatties(int numberOfBurgers) {
    return numberOfBurgers * 2;  // Calculate and return in one step
}
```

### Using Return Values

When a method returns a value, you can use it in several ways:

**1. Store in a variable:**
```java
int pattiesNeeded = countPatties(5);
System.out.println("Need " + pattiesNeeded + " patties");
```

**2. Use directly in an expression:**
```java
int totalPatties = countPatties(5) + countPatties(3);  // 10 + 6 = 16
```

**3. Pass to another method:**
```java
System.out.println(countPatties(5));  // Prints: 10
```

**4. Use in a condition:**
```java
if (countPatties(orders) > 50) {
    System.out.println("Big order! Call for backup!");
}
```

The method call is essentially replaced by its return value.

### `void` vs Return Types

| Aspect | `void` Methods | Methods with Return Types |
|--------|----------------|---------------------------|
| **Purpose** | Perform an action | Compute and return a value |
| **Return type** | `void` | `int`, `double`, `String`, `boolean`, etc. |
| **Return statement** | Optional (or `return;` with no value) | Required with matching type |
| **How to call** | `doSomething();` | `result = calculateSomething();` |
| **Example** | `printReceipt()` | `calculateTotal()` |
| **Analogy** | "Clean the grill" (action) | "Count the customers" (gives information) |

**Common mistake:** Trying to use a `void` method's "value":

```java
int x = printMessage();  // ERROR! void methods don't return anything
```

##  Variable Scope


**Scope** refers to where in your code a variable can be accessed. Think of it like Bob's kitchen versus the dining room:

- Bob's secret recipes stay in the kitchen‚Äîcustomers can't see them
- The menu prices are visible in the dining room
- What happens in the kitchen stays in the kitchen!

In Java, variables declared inside a method are **local** to that method. They don't exist outside of it, and other methods can't see them.

### Local Variables

A **local variable** is declared inside a method and only exists while that method is running:

```java
public static void methodOne() {
    int total = 100;  // Local to methodOne
    System.out.println(total);  // Works fine
}

public static void methodTwo() {
    int total = 200;  // Different variable! Local to methodTwo
    System.out.println(total);  // Works fine, prints 200
}

public static void main(String[] args) {
    methodOne();  // Prints 100
    methodTwo();  // Prints 200
    // System.out.println(total);  // ERROR! 'total' doesn't exist here
}
```

Even though both methods have a variable named `total`, they're completely separate‚Äîlike two different order pads in two different restaurants.

### Parameters are Local

Parameters are also local variables‚Äîthey exist only inside their method. Importantly, for primitives and Strings, the parameter is a **copy** of the argument:

```java
public static void tryToChange(int value) {
    value = 999;  // Only changes the local copy!
    System.out.println("Inside method: " + value);  // 999
}

public static void main(String[] args) {
    int original = 42;
    tryToChange(original);
    System.out.println("After method: " + original);  // Still 42!
}
```

The method gets a copy of `42`, not the original variable. Changing the copy doesn't affect `original` in `main`. It's like writing a number on a sticky note and handing it to someone‚Äîif they cross it out and write a new number, your original note is unchanged.

### Flowchart: Scope Visualization

````python
# @title
mm("""
flowchart TB
    subgraph main["main() scope"]
        A["int original = 42"]
        B["calls tryToChange(original)"]
        C["original is still 42"]
    end

    subgraph method["tryToChange() scope"]
        D["int value = 42 (copy)"]
        E["value = 999"]
        F["method ends, value disappears"]
    end

    A --> B
    B -->|"copy value 42"| D
    D --> E --> F
    F -->|"return to main"| C""")
````

**Scope Visualization:** Each method has its own isolated workspace. Values are copied in, and the return value is copied out. Changes to local variables don't escape their method.

This is actually a good thing! It means methods can't accidentally mess up each other's data. Each method is self-contained.

##  Method Overloading


Bob wants different ways to take an order:
- Just a burger name: `"Burger of the Day"`
- Burger and quantity: `"Burger of the Day"`, `3`
- Burger, quantity, and special instructions: `"Burger of the Day"`, `3`, `"no onions"`

Without overloading, you'd need different method names:

```java
makeOrderSimple("Burger of the Day");
makeOrderWithQuantity("Burger of the Day", 3);
makeOrderWithInstructions("Burger of the Day", 3, "no onions");
```

That's messy! There has to be a better way...

### What is Overloading?

**Method overloading** lets you define multiple methods with the **same name** but **different parameter lists**. Java figures out which one to call based on the arguments you provide.

```java
public static void makeOrder(String burger) { ... }
public static void makeOrder(String burger, int qty) { ... }
public static void makeOrder(String burger, int qty, String special) { ... }
```

Now you can call:

```java
makeOrder("Cheeseburg");                      // Calls first version
makeOrder("Cheeseburg", 2);                   // Calls second version
makeOrder("Cheeseburg", 2, "extra pickles"); // Calls third version
```

It's like Bob's regular customers. Teddy says "the usual" and Bob knows exactly what that means for Teddy. Mort says "the usual" and Bob makes something different. Same words, different actions based on context.

### Overloading Rules

For methods to be valid overloads, they must differ in their **parameter lists**:

| Difference | Valid Overload? | Example |
|------------|-----------------|--------|
| Different **number** of parameters | ‚úì Yes | `print()` vs `print(String msg)` |
| Different **types** of parameters | ‚úì Yes | `calc(int x)` vs `calc(double x)` |
| Different **order** of types | ‚úì Yes | `f(int a, String b)` vs `f(String a, int b)` |
| Different **parameter names** only | ‚úó No | `f(int x)` vs `f(int y)` ‚Äî same signature! |
| Different **return type** only | ‚úó No | `int f()` vs `String f()` ‚Äî ambiguous! |

**Key insight:** Java uses the **method signature** (name + parameter types) to distinguish methods. Return type is NOT part of the signature.

**Full Program #4: OverloadedOrders.java** shows three overloaded versions of `makeOrder()`:
- One parameter: just the burger
- Two parameters: burger + quantity
- Three parameters: burger + quantity + special instructions

Java automatically selects the right version based on how many arguments you pass and their types. This makes the API cleaner and more flexible!

````python
%%writefile OverloadedOrders.java
public class OverloadedOrders {

    // Version 1: Just the burger name (quantity defaults to 1)
    public static void makeOrder(String burger) {
        System.out.println("Order: 1x " + burger);
        System.out.println("  (Standard preparation)");
        System.out.println();
    }

    // Version 2: Burger name and quantity
    public static void makeOrder(String burger, int quantity) {
        System.out.println("Order: " + quantity + "x " + burger);
        System.out.println("  (Standard preparation)");
        System.out.println();
    }

    // Version 3: Burger, quantity, and special instructions
    public static void makeOrder(String burger, int quantity, String special) {
        System.out.println("Order: " + quantity + "x " + burger);
        System.out.println("  Special: " + special);
        System.out.println();
    }

    public static void main(String[] args) {
        System.out.println("=== Bob's Burgers Order System ===\n");

        // Java picks the right method based on arguments!
        makeOrder("Papaya Was a Rolling Stone Burger");

        makeOrder("Do the Brussel Burger", 3);

        makeOrder("Bet It All On Black Garlic Burger", 2, "no tomato, extra cheese");
    }
}
````

### Method Chaining

Since String methods return Strings, you can **chain** them together:

Read left to right:

```java
String original = "bob";

```

This is a common bug! Always remember: if you want the changed version, you must save it.

## Introduction to Recursion

**Recursion** is when a method calls itself. Yes, really!

Think of Louise Belcher's schemes. She often has a scheme within a scheme within a scheme. Each scheme might require completing a smaller scheme first, until eventually you reach a simple action you can just do.

```
Louise's Master Plan:
  ‚îî‚îÄ‚îÄ Scheme A (requires completing Scheme B first)
        ‚îî‚îÄ‚îÄ Scheme B (requires completing Scheme C first)
              ‚îî‚îÄ‚îÄ Scheme C (simple enough to just do!)
```

Recursion works the same way: a problem is solved by breaking it into smaller versions of itself, until you reach a version simple enough to solve directly.

### The Two Requirements

Every recursive method MUST have:

**1. Base Case** ‚Äî The simple condition where we stop recursing and return a direct answer. Without this, you get infinite recursion (like Louise's schemes spiraling out of control)!

**2. Recursive Case** ‚Äî The method calls itself with a "smaller" or "simpler" version of the problem. Each recursive call must move toward the base case.

```java
public static void recursiveMethod(parameters) {
    if (baseCondition) {
        // Base case: stop recursing
        return something;
    } else {
        // Recursive case: call self with simpler input
        return recursiveMethod(smallerProblem);
    }
}
```

**Warning:** Forgetting the base case or not moving toward it causes **stack overflow**‚Äîthe program runs out of memory tracking all the nested calls.

### Classic Example: Countdown

Let's count down from n to 1, then say "Burgers ready!"

```java
public static void countdown(int n) {
    if (n <= 0) {
        // Base case: we've counted down to zero
        System.out.println("Burgers ready!");
    } else {
        // Recursive case: print current number, then count down from n-1
        System.out.println(n);
        countdown(n - 1);  // Call ourselves with a smaller number
    }
}
```

Trace through `countdown(3)`:
1. `countdown(3)`: prints 3, calls `countdown(2)`
2. `countdown(2)`: prints 2, calls `countdown(1)`
3. `countdown(1)`: prints 1, calls `countdown(0)`
4. `countdown(0)`: base case! prints "Burgers ready!" and returns
5. Each call returns back up the chain

Output: `3`, `2`, `1`, `Burgers ready!`

### Recursion Flowchart

````python
# @title
mm("""
flowchart TD
    A(["countdown(3)"]) --> B{"Is 3 <= 0?"}
    B -->|No| C["Print 3"]
    C --> D(["countdown(2)"])
    D --> E{"Is 2 <= 0?"}
    E -->|No| F["Print 2"]
    F --> G(["countdown(1)"])
    G --> H{"Is 1 <= 0?"}
    H -->|No| I["Print 1"]
    I --> J(["countdown(0)"])
    J --> K{"Is 0 <= 0?"}
    K -->|Yes| L["Print 'Burgers ready!'"]
    L --> M["Return"]
    M -.->|"returns to"| G
    G -.->|"returns to"| D
    D -.->|"returns to"| A
""")
````

**Recursion Flowchart:** Notice how each call waits for the one it spawned to complete. The calls "stack up" and then "unwind" as each one returns. This is why it's called the "call stack"!

### Recursion with Return Values

Recursion really shines when computing values. The classic example is **factorial**:

- 5! = 5 √ó 4 √ó 3 √ó 2 √ó 1 = 120
- Recursively: 5! = 5 √ó 4!

```java
public static int factorial(int n) {
    if (n <= 1) {
        return 1;  // Base case: 0! = 1, 1! = 1
    } else {
        return n * factorial(n - 1);  // n! = n * (n-1)!
    }
}
```

Trace `factorial(4)`:
1. `factorial(4)` returns `4 * factorial(3)` ‚Üí waiting...
2. `factorial(3)` returns `3 * factorial(2)` ‚Üí waiting...
3. `factorial(2)` returns `2 * factorial(1)` ‚Üí waiting...
4. `factorial(1)` returns `1` ‚Üê base case!
5. Now unwind: `2 * 1 = 2`, then `3 * 2 = 6`, then `4 * 6 = 24`

Final result: 24

**Full Program #6: RecursiveBurgers.java** demonstrates three recursive methods:
- `orderCountdown()` ‚Äî void recursive method with printing
- `calculateBulkDiscount()` ‚Äî recursive calculation with multiple base cases
- `sumUpTo()` ‚Äî classic recursive sum

Notice how each has a clear base case that stops the recursion.

````python
%%writefile RecursiveBurgers.java
public class RecursiveBurgers {

    // Recursive countdown for order ready announcement
    public static void orderCountdown(int seconds) {
        if (seconds <= 0) {
            System.out.println("ORDER UP!");
        } else {
            System.out.println(seconds + "...");
            orderCountdown(seconds - 1);
        }
    }

    // Recursive discount calculation
    // Every 10 burgers adds 5% discount, max 25%
    public static int calculateBulkDiscount(int burgers) {
        if (burgers < 10) {
            return 0;  // Base case: less than 10, no discount
        } else if (burgers >= 50) {
            return 25;  // Cap at 25%
        } else {
            // Add 5% for this group of 10, plus discount for remaining
            return 5 + calculateBulkDiscount(burgers - 10);
        }
    }

    // Recursive sum: add up all numbers from 1 to n
    public static int sumUpTo(int n) {
        if (n <= 0) {
            return 0;  // Base case
        } else {
            return n + sumUpTo(n - 1);  // n + sum of (1 to n-1)
        }
    }

    public static void main(String[] args) {
        System.out.println("=== Order Countdown ===");
        orderCountdown(5);

        System.out.println("\n=== Bulk Discounts ===");
        System.out.println("5 burgers: " + calculateBulkDiscount(5) + "% off");
        System.out.println("15 burgers: " + calculateBulkDiscount(15) + "% off");
        System.out.println("35 burgers: " + calculateBulkDiscount(35) + "% off");
        System.out.println("100 burgers: " + calculateBulkDiscount(100) + "% off (capped)");

        System.out.println("\n=== Sum Calculation ===");
        System.out.println("Sum 1 to 5: " + sumUpTo(5));   // 15
        System.out.println("Sum 1 to 10: " + sumUpTo(10)); // 55
    }
}
````

##  Putting It All Together


How do you know when to create a method? Here are some guidelines:

**Create a method when:**

1. **Code repeats** ‚Äî If you're copying and pasting, make it a method
2. **Distinct task** ‚Äî If code does one specific thing, give it a name
3. **Cleaner main** ‚Äî If `main` is getting long, break it up
4. **Reusability** ‚Äî If you might need this logic elsewhere
5. **Testability** ‚Äî Smaller methods are easier to test and debug

Think of Bob's kitchen organization:
- Prep station (preparing ingredients) ‚Äî one method
- Grill station (cooking patties) ‚Äî another method
- Assembly station (building burgers) ‚Äî another method
- Register (calculating totals) ‚Äî another method

Each station has a clear responsibility. Your methods should too!

**Full Program #7: BobsBurgerShop.java** brings everything together:
- **Void methods** for greetings and display (`greetCustomer`, `printReceipt`)
- **Returning methods** for calculations (`getBurgerPrice`, `calculateSubtotal`)
- **Boolean parameters** for conditional logic (`applyLoyaltyDiscount`)
- **Switch statement** for multi-way pricing
- **Method chaining** ‚Äî results flow from one method to the next

Notice how clean `main` is! It reads almost like English: greet the customer, print the special, calculate prices, print receipt. The details are hidden inside well-named methods.

````python
%%writefile BobsBurgerShop.java
public class BobsBurgerShop {

    // ===== GREETING METHODS =====
    public static void greetCustomer(String name) {
        System.out.println("\n" + "=".repeat(40));
        System.out.println("Welcome to Bob's Burgers, " + name + "!");
        System.out.println("=".repeat(40));
    }

    // ===== MENU METHODS =====
    public static void printBurgerOfTheDay() {
        System.out.println("\n*** TODAY'S SPECIAL ***");
        System.out.println("The 'If Looks Could Kale' Burger - $6.50");
        System.out.println("A beef patty with kale slaw and garlic aioli");
    }

    // ===== PRICING METHODS =====
    public static double getBurgerPrice(String burger) {
        switch (burger.toLowerCase()) {
            case "burger of the day":
                return 6.50;
            case "cheeseburger":
                return 5.50;
            case "hamburger":
                return 4.95;
            default:
                return 5.95;  // Standard price
        }
    }

    public static double calculateSubtotal(String burger, int quantity) {
        return getBurgerPrice(burger) * quantity;
    }

    public static double calculateTax(double amount) {
        return amount * 0.08;  // 8% tax
    }

    public static double applyLoyaltyDiscount(double amount, boolean isLoyal) {
        if (isLoyal) {
            return amount * 0.90;  // 10% off for loyal customers
        }
        return amount;
    }

    // ===== DISPLAY METHODS =====
    public static String formatMoney(double amount) {
        return String.format("$%.2f", amount);
    }

    public static void printReceipt(String customer, String burger, int qty,
                                    double subtotal, double discount,
                                    double tax, double total) {
        System.out.println("\n" + "-".repeat(30));
        System.out.println("BOB'S BURGERS - RECEIPT");
        System.out.println("-".repeat(30));
        System.out.println("Customer: " + customer);
        System.out.println(qty + "x " + burger);
        System.out.println("Subtotal: " + formatMoney(subtotal));
        if (discount > 0) {
            System.out.println("Loyalty Discount: -" + formatMoney(discount));
        }
        System.out.println("Tax: " + formatMoney(tax));
        System.out.println("-".repeat(30));
        System.out.println("TOTAL: " + formatMoney(total));
        System.out.println("-".repeat(30));
        System.out.println("Thanks for eating at Bob's!");
    }

    // ===== MAIN =====
    public static void main(String[] args) {
        // Process an order for Teddy (loyal customer)
        String customer = "Teddy";
        String burger = "Burger of the Day";
        int quantity = 2;
        boolean isLoyalCustomer = true;

        greetCustomer(customer);
        printBurgerOfTheDay();

        // Calculate pricing step by step
        double subtotal = calculateSubtotal(burger, quantity);
        double afterDiscount = applyLoyaltyDiscount(subtotal, isLoyalCustomer);
        double discountAmount = subtotal - afterDiscount;
        double tax = calculateTax(afterDiscount);
        double total = afterDiscount + tax;

        // Print the receipt
        printReceipt(customer, burger, quantity, subtotal,
                     discountAmount, tax, total);
    }
}
````

### Common Mistakes

Watch out for these frequent method errors:

| Mistake | What Happens | Fix |
|---------|--------------|-----|
| Forgetting `return` | Compile error: "missing return statement" | Add `return value;` at end of all paths |
| Wrong return type | Compile error: "incompatible types" | Match return type to what you return |
| Wrong parameter order | Wrong values used, unexpected behavior | Match argument order to parameter order |
| Calling without `()` | Compile error or reference instead of call | Always use `methodName()` |
| Using void in expression | Compile error: "void cannot be converted" | Void methods don't return usable values |
| Infinite recursion | StackOverflowError, program crashes | Ensure base case is reached |
| Not saving return value | Value is computed but lost | Store in variable: `int x = method();` |
| Defining inside main | Compile error | Define methods outside main, inside class |

### Key Takeaways

1. **Methods are reusable recipes** ‚Äî Define once, call many times. They reduce repetition and organize code.

## Glossary

| Term | Definition |
|------|------------|
| **Method** | A named, reusable block of code that performs a specific task |
| **Parameter** | A variable in a method definition that receives input values |
| **Argument** | The actual value passed to a method when calling it |
| **Return Type** | The data type of value a method gives back (or `void` for none) |
| **`void`** | Keyword indicating a method doesn't return a value |
| **`return`** | Statement that exits a method and optionally sends back a value |
| **Calling/Invoking** | Executing a method by using its name with parentheses |
| **Local Variable** | A variable declared inside a method, only accessible within it |
| **Scope** | The region of code where a variable can be accessed |
| **Method Overloading** | Defining multiple methods with the same name but different parameters |
| **Method Signature** | The combination of method name and parameter types |
| **Recursion** | A technique where a method calls itself to solve a problem |
| **Base Case** | The condition in recursion where the method stops calling itself |
| **Recursive Case** | The part of recursion where the method calls itself with a simpler input |
| **Stack Overflow** | Error from infinite recursion; too many nested method calls |
| **Method Chaining** | Calling methods on the results of other methods in sequence |
| **Immutable** | Cannot be changed; String methods return new strings, not modified originals |
| **`static`** | Keyword indicating a method belongs to the class, not an instance |
| **DRY Principle** | "Don't Repeat Yourself" ‚Äî use methods to avoid code duplication |
